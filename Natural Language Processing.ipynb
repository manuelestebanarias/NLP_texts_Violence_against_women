## Finding word stem

```python
def stem( word):
    for suffix in ['ing','ly','ed','ious','ies', 'ive','es','s','ment']:
        if word.endswith(suffix):
            return word[:-len(suffix)]
    return word
```

```python
#for identifaying suffix
re.findall(r'^.*(ing|ly|ed|ious|ies|ive|es|s|ment)$','processing')
```

```python
#but we prefere  to split a word into stem and suffix
re.findall(r'^(.*)(ing|ly|ed|ious|ies|ive|es|s|ment)$','processing')
```

```python
#the ? force the selection
re.findall(r'^(.*?)(ing|ly|ed|ious|ies|ive|es|s|ment)$','processes')
```

```python
#wat if there is no suffix? --> a second ?
re.findall(r'^(.*?)(ing|ly|ed|ious|ies|ive|es|s|ment)?$','languaje')
```

```python
#improve the funciton
def stem( word):
    regexp= r'^(.*?)(ing|ly|ed|ious|ies|ive|es|s|ment)?$'
    stem,suffix= re.findall(regexp,word)[0]
    return stem
```

```python
class IndexedText(object):
    def __init__(self,stemmer,text):
        self._text = text
        self._stemmer = stemmer
        self._index = nltk.Index((self._stem(word),i)
                                for (i,word) in enumerate (text))
    def concordance(self,word,width=40):
        key = self._stem(word)
        wc=int(width/4)
        for i in self._index[key]:
            lcontent = ' '.join(self._text[i-wc:i])
            rcontent = ' '.join(self._text[i:i+wc])
            ldisplay='%*s'%(width,lcontent[-width:])
            rdisplay='%-*s'%(width,rcontent[:width])
            print(ldisplay,rdisplay)
    def _stem(self, word):
        return self._stemmer.stem(word).lower()

porter =nltk.PorterStemmer()
grail=nltk.corpus.webtext.words('grail.txt')
text= IndexedText(porter, grail)
#grail.concordance('lie')
text.concordance('lie')
```

## Representing tagged corporas

```python
tagged_token=nltk.tagged.str2uple('fly/NN')
tagged_token
```

```python
#find the most frequent noun tags?
def findtags(tag_prefix,tagged_text):
    cfd=nltk.ConditionalFreqDist((tag,word) for (word, tag) in tagged_text
                                if tag.startswith(tag_prefix))
    return dict((tag,list(cfd[tag].keys())[:5]) for tag in cfd.conditions())
tagdict=findtags('NN', nltk.corpus.brown.tagged_words(categories='news'))
for tag in sorted(tagdict):
    print(tag,tagdict[tag])
```

```python
# in terms of pos tag?
brown_lrnd_text=nltk.corpus.brown.tagged_words(categories='learned')
tags=[b[1] for (a,b) in nltk.bigrams(brown_lrnd_text) if a[0]=='often']
fd=nltk.FreqDist(tags)
fd.tabulate()
```
